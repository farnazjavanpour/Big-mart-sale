```python
dummydata=pd.get_dummies(cdata, columns=['workclass', 'education', 'marital_status', 'occupation'])
dummydata.head()
```

```python
def income_to_numeric(x):
    if x=='>50K':
        return 1
    if x=='<=50K':
        return 0
    
cdata['income_bracket'] = cdata['income_bracket'].apply(income_to_numeric)
def sex_to_numeric(x):
    if x=='Male':
        return 1
    if x=='Female':
        return 0
    
cdata['gender'] = cdata['gender'].apply(sex_to_numeric)
```

```python
print(cdata['income_bracket'])
```

```python
x=dummydata.drop(['income_bracket','gender'],axis=1)
y=dummydata['income_bracket']
```

```python
x=(x-x.min())/(x.max()-x.min())
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_std=StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
```

```python
prinicipaleComponents=pca.fit_transform(x_std)
```

```python
principleDf=pd.DataFrame(data=prinicipaleComponents)
```

```python
principleDf
```

```python
from sklearn.model_selection import train_test_split
principalDF_train,principalDF_test,Y_train,Y_test=train_test_split(principleDf,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(principalDF_train,Y_train)
```

```python
pred=model.predict(principalDF_test)
from sklearn import metrics
pca_acc=metrics.accuracy_score(Y_test,pred)*100
pca_acc
```
